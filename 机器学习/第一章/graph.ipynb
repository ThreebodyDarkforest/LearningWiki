```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10)
y = 1 / (1 + np.e ** (-x))
note = plt.subplot()

note.annotate(r'$y = \dfrac{1}{1 + e^z}$', 
              xy = (-2.5, 1 / (1 + np.e ** (2.5))),
              xytext = (-7, 0.5),
              arrowprops = dict(facecolor = 'black', shrink = 0.05, width = 1, headwidth = 8))

plt.plot(x, y)
plt.show()
```
